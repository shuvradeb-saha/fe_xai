```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

from pyexplainer import pyexplainer_pyexplainer
```

```python
# Load dataset
wine_data = load_iris(as_frame=True)

print(wine_data.feature_names)

df = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)

# df = df.rename(columns={'od280/od315_of_diluted_wines': 'od280_od315_diluted_wines'})

df["target"] = wine_data.target

# data = wine_data.data
# target = wine_data.target
# target_names = wine_data.target_names
# feature_names = wine_data.feature_names

X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns="target"), df["target"], test_size=0.2
)

display(X_test)


# # Convert to pandas DataFrame
# df = pd.DataFrame(data, columns=feature_names)
# df['target'] = target
# df.head(5)
```

```python
# Prepare the model
rf = RandomForestClassifier(n_estimators=50, random_state=42)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)

# Evaluation
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

### Explain using pyexplainer. We need pandas dataframe to work with the pyexplainer.
To initialize pyExplainer we need train dataset (X_train), the result columns (y_train), independent variable columns or fetaure names as pandas Index, dependent variable column name or label and the blackbox model that we have trained on the above. 

```python
# Initialize the PyExplainer
independent = X_train.columns
print(independent)
depentent = "target"
py_explainer = pyexplainer_pyexplainer.PyExplainer(
    X_train=X_train,
    y_train=y_train,
    indep=independent,
    dep=depentent,
    blackbox_model=rf,
)
```

```python
X_explain = X_test.iloc[[0]]
y_explain = y_test.iloc[[0]]

display(type(X_explain))
display(type(y_explain))
# Explore the explainer
rules = py_explainer.explain(
    X_explain=X_explain, y_explain=y_explain, search_function="CrossoverInterpolation"
)
```
