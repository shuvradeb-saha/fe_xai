```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
from IPython.display import display
```

```python
cancer_data = load_breast_cancer()
df = pd.DataFrame(cancer_data.data, columns=cancer_data.feature_names)
df["target"] = cancer_data.target
pd.set_option("display.max_columns", None)
print(df.shape)

print("Cancer detected:")
display(df[df["target"] == 1].head(5))
print("No cancer:")
display(df[df["target"] == 0].head(5))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the distribution of the target variable
sns.countplot(x='target', data=df)
plt.title('Distribution of Target Variable')
plt.xlabel('Target')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the pairplot of the first few features
sns.pairplot(df, vars=cancer_data.feature_names[:4], hue='target', markers=["o", "s"])
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
# Separate target from the features
data = df.drop(columns="target")
target = df["target"]

# Prepare training and test set
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.2, random_state=15
)

display(X_train.head(1))

rf = RandomForestClassifier(n_estimators=50)
rf.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = rf.predict(X_test)
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

### Use dalex to explain the result now.

```python
# We have the model. Now prepare the data for explaining with dalex
import dalex as dx

explainer = dx.Explainer(rf, X_train, y_train, label="RFC Cancer")
```

```python
# Model performance
mp = explainer.model_performance()
display(mp)
# mp.result
mp.plot()
```

```python
# Feature Importance: feature contribution to the model
explainer.model_parts()
```

```python
# Feature importance plot
explainer.model_parts().plot()
```
